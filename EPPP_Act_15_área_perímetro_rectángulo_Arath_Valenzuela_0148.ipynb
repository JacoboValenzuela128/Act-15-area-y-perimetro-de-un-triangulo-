El código sigue un enfoque modular y procedural, lo que significa que cada tarea se delega a funciones independientes, facilitando la lectura, mantenimiento y comprensión del flujo de ejecución sin necesidad de utilizar clases ni objetos.
* Para los cálculos geométricos se aplican las siguientes fórmulas matemáticas básicas:
1. Área del rectángulo:
$A=b×h$
2. Perímetro del rectángulo:
$P=2⋅(b+h)$
* Donde b representa la medida de la base y h la medida de la altura de la figura.

* El comportamiento del programa se controla mediante la función $menu()$, la cual despliega las opciones disponibles al usuario.
* Un ciclo while mantiene el programa en ejecución de forma continua hasta que se seleccione la opción de salida.
* Dentro del ciclo, las estructuras condicionales if, elif y else interpretan la elección del usuario y dirigen la ejecución hacia la función correspondiente.
* Tanto $arearectangulo()$ como $perimetrorectangulo()$ solicitan los valores de base y altura en el momento de ser llamadas, realizan el cálculo matemático y muestran el resultado en pantalla con un formato limpio.
* No se incluye validación de entrada de datos para mantener el código enfocado exclusivamente en la lógica de control y llamadas a funciones. La interfaz utiliza emojis para hacer la interacción más visual y amigable.

```python
def arearectangulo():
    """Calcula y muestra el área de un rectángulo."""
    print("\n📐 Cálculo del Área del Rectángulo")
    base = float(input("📥 Ingresa la base del rectángulo: "))
    altura = float(input("📥 Ingresa la altura del rectángulo: "))
    area = base * altura
    print(f"✅ El área del rectángulo es: {area:.2f} unidades²")

def perimetrorectangulo():
    """Calcula y muestra el perímetro de un rectángulo."""
    print("\n📏 Cálculo del Perímetro del Rectángulo")
    base = float(input("📥 Ingresa la base del rectángulo: "))
    altura = float(input("📥 Ingresa la altura del rectángulo: "))
    perimetro = 2 * (base + altura)
    print(f"✅ El perímetro del rectángulo es: {perimetro:.2f} unidades")

def menu():
    """Muestra el menú interactivo y controla el flujo del programa."""
    opcion = 0
    while opcion != 3:
        print("\n🔹 ══════ MENÚ PRINCIPAL ══════ 🔹")
        print("1️⃣  Calcular área del rectángulo")
        print("2️⃣  Calcular perímetro del rectángulo")
        print("3️⃣  Salir del programa")
        print("═══════════════════════════════")

        opcion = int(input("👉 Elige una opción (1-3): "))

        if opcion == 1:
            arearectangulo()
        elif opcion == 2:
            perimetrorectangulo()
        elif opcion == 3:
            print("\n👋 ¡Gracias por usar el programa! Hasta pronto. ✨")
        else:
            print("⚠️ Opción no válida. Por favor, selecciona un número del 1 al 3.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
```
